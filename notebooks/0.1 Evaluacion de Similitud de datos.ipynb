**Calidad de balanceos en strings:**

```python
## imprimir los strings que estan en data de entrenamienot y testeo sin balancear pero que no estan en el escaler cuando se entreneo el modelo
keys_not_in = []
for df_temp in [x_train,x_test]:
    for col in scaler.reindex_OneHotEncoding.columns[1:]:
        strings = df_temp[col].drop_duplicates().values.tolist()
        for strr in strings:
            if strr not in scaler.reindex_OneHotEncoding[col].drop_duplicates().values.tolist():
                keys_not_in.append(str(col)+"_"+str(strr))
                print("Datos no existen: ",str(col)+"_"+str(strr))
    print("----------------------------------------------------")
keys_not_in= list(set(keys_not_in))
keys_not_in
```

```python
## extraer la data de balanceo del mejor modelo
try:
    llave = info_save_select['modelo_produccion']['name_model']
    print(llave)
    try:
        extract = info_save_all[int(llave[-2:])]
    except:
        extract = info_save_all[int(llave[-1:])]
    for balance_data in [extract["X_balance"],extract["X_train"]]: 
        for col in keys_not_in:
            if col not in balance_data.columns:
                print("No esta: ", col)
        print("----------------------------")
except:
    print(llave)
```

```python
risk = 0
true__ = 0
for col in  train_test_interes.columns.tolist():
    try:
        mini = train_test_interes[col].min()
        maxi = train_test_interes[col].max()
    
        sig = train_test_interes[col].std()
        mean = train_test_interes[col].mean()
        
        mini2 = read_interes[col].min()
        maxi2 = read_interes[col].max()
    
        std2 = (read_interes[col] - mean)/sig
        if mini2 < mini:
            print("Min Fuera de rango:")
            risk = risk + (read_interes[col] <mini).sum()
            true__ = true__ + read[read_interes[col] <mini][params['variable_apertura']].sum()
            print(col, "Umbral: ",mini)
            print(col,".Valor: ",np.round(mini2,2),". N= ",(read_interes[col] <mini).sum(),"%:",np.round(100*(read_interes[col] <mini).sum()/read_interes.shape[0],2))
        if maxi2 > maxi:
            print("Max Fuera de rango:")
            risk = risk + (read_interes[col] >maxi).sum()
            true__ = true__ + read[read_interes[col] > maxi][params['variable_apertura']].sum()
            print(col, "Umbral: ",maxi)
            print(col,".Valor: ",np.round(maxi2,2),". N= ",(read_interes[col] >maxi).sum(),"%:",np.round(100*(read_interes[col] >maxi).sum()/read_interes.shape[0],2))
    except:
        print("Error",col)
    finally:
        print("----------------------------")
```
