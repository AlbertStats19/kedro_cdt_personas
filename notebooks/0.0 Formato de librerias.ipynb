# Actualizacion de librerias

## Crear el archivo desde la terminal:

**python -m pip list > src/requirements_mvp8.txt**

## Leer ese archivo:

```python
ruta_archivo = '/home/sagemaker-user/data-bbog-integration-fabrica-personas/src/requirements_mvp8.txt'
```

```python
# leyendo el archivo

# Abre el archivo con los paquetes
with open(ruta_archivo, 'r') as file:
    lines = file.readlines()
```

## Corregir el archivo

```python
ruta_archivo1 = '/home/sagemaker-user/data-bbog-integration-fabrica-personas/src/requirements_mvp8.txt'
```

```python
# Guarda el archivo
with open(ruta_archivo1, 'w') as out_file:
    for line in lines:
        if line[:len('Package')] == 'Package':
            print(f'Ignorando: {line}')
        else:
            # Se ignoran las líneas de encabezado o separadores
            if '----------------------------' in line or line.strip() == '':
                continue
            
            # Se separan el nombre del paquete y la versión
            parts = line.split()
            
            if len(parts) >= 2:
                package_name = parts[0]
                version = parts[1]
                out_file.write(f"{package_name}=={version}\n")

print("Actualizado correctamente.")
```
