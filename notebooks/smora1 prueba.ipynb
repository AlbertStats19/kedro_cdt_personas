```python
def homologate_region(df: pd.DataFrame,parametros: Dict) -> pd.DataFrame:
    """
    Homologa las regiones en el DataFrame en base al diccionario de homologaciones.

    Args:
        df (pd.DataFrame): DataFrame que contiene la columna de departamentos a homologar.
        departamento_col (str): Nombre de la columna que contiene los departamentos.

    Returns:
        pd.DataFrame: DataFrame con la nueva columna de regiones homologadas.
    """
    
    # Diccionario de homologaci√≥n de departamentos a regiones
    homologations = {
        "amazonas": "Region_Amazonica",
        "antioquia": "Region_Andina",
        "arauca": "Region_Orinoquia",
        "archipielago_de_san_andres_providencia_y": "Region_Caribe",
        "atlantico": "Region_Caribe",
        "bogota_d_c": "Region_Central",
        "bolivar": "Region_Caribe",
        "boyaca": "Region_Andina",
        "caldas": "Region_Andina",
        "caqueta": "Region_Amazonica",
        "casanare": "Region_Orinoquia",
        "cauca": "Region_Pacifica",
        "cesar": "Region_Caribe",
        "choco": "Region_Pacifica",
        "cordoba": "Region_Caribe",
        "cundinamarca": "Region_Andina",
        "guainia": "Region_Amazonica",
        "guaviare": "Region_Amazonica",
        "huila": "Region_Andina",
        "la_guajira": "Region_Caribe",
        "magdalena": "Region_Caribe",
        "meta": "Region_Orinoquia",
        "narino": "Region_Pacifica",
        "norte_de_santander": "Region_Andina",
        "putumayo": "Region_Amazonica",
        "quindio": "Region_Andina",
        "risaralda": "Region_Andina",
        "santander": "Region_Andina",
        "sucre": "Region_Caribe",
        "tolima": "Region_Andina",
        "valle_del_cauca": "Region_Pacifica",
        "vaupes": "Region_Amazonica",
        "vichada": "Region_Orinoquia"
    }
    
    departamento_col= parametros['departamento_col']
    # Normalizar los nombres de los departamentos (ej. a min√∫sculas y sin espacios)
    df[departamento_col] = df[departamento_col].str.lower().str.replace(r'\s+', '_', regex=True)
    
    # Mapear los departamentos a regiones usando el diccionario de homologaciones
    df["region"] = df[departamento_col].map(homologations).fillna("Other_Region")
    
    return df
```

```python
def preprocesar_feature_df(df: pd.DataFrame, params: Dict[str, Any]) -> pd.DataFrame:
    """
    Preprocesa el DataFrame para codificar variables categ√≥ricas y realizar imputaci√≥n de valores faltantes.

    Parameters
    ----------
    df : pd.DataFrame
        DataFrame con las caracter√≠sticas originales.
    params : Dict[str, Any]
        Diccionario de par√°metros que incluye la columna a conservar.

    Returns
    -------
    pd.DataFrame
        DataFrame preprocesado con variables categ√≥ricas codificadas y valores faltantes imputados.
    """
    logger.info("Iniciando el preprocesamiento del DataFrame...")

    # Identificar columnas categ√≥ricas
    categorical_columns = df.select_dtypes(include=['object']).columns.tolist()
    columna_conservar = params['columna_conservar']
    
    # Codificaci√≥n de columnas categ√≥ricas
    df_encoded = df.copy()
    label_encoders = {}
    for col in categorical_columns:
        le = LabelEncoder()
        df_encoded[col] = le.fit_transform(df_encoded[col].astype(str))
        label_encoders[col] = le

    # Imputar valores faltantes
    for col in df_encoded.columns:
        if df_encoded[col].dtype == 'object':
            df_encoded[col].fillna(df_encoded[col].mode()[0], inplace=True)
        else:
            df_encoded[col].fillna(df_encoded[col].median(), inplace=True)

    # Convertir todos los datos a num√©ricos
    df_encoded = df_encoded.apply(pd.to_numeric, errors='coerce')

    return df_encoded
```

```python
def separar_caracter√≠sticas(df: pd.DataFrame, columna_conservar: str) -> Tuple[pd.DataFrame, pd.Series]:
    """
    Separa las caracter√≠sticas y la variable objetivo del DataFrame.

    Parameters
    ----------
    df : pd.DataFrame
        DataFrame preprocesado.
    columna_conservar : str
        Nombre de la columna que se utilizar√° como variable objetivo.

    Returns
    -------
    Tuple[pd.DataFrame, pd.Series]
        DataFrame con caracter√≠sticas y Serie con la variable objetivo.
    """
    logger.info("Separando caracter√≠sticas y la variable objetivo...")

    X = df.drop(columns=[columna_conservar])
    y = df[columna_conservar]
    
    return X, y
```

```python
from xgboost import XGBRegressor

def calcular_importancia(X: pd.DataFrame, y: pd.Series) -> pd.DataFrame:
    """
    Calcula la importancia de las caracter√≠sticas utilizando un modelo de XGBoost.

    Parameters
    ----------
    X : pd.DataFrame
        DataFrame con las caracter√≠sticas.
    y : pd.Series
        Serie con la variable objetivo.

    Returns
    -------
    pd.DataFrame
        DataFrame con la importancia de las caracter√≠sticas ordenadas.
    """
    logger.info("Calculando la importancia de las caracter√≠sticas usando XGBoost...")

    # Definir el modelo XGBoost
    regressor = XGBRegressor(max_depth=5, random_state=42, n_estimators=60)
    
    # Ajustar el modelo a los datos
    regressor.fit(X, y)
    
    # Obtener la importancia de las caracter√≠sticas
    importances = regressor.feature_importances_
    feature_names = X.columns
    importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
    
    # Ordenar las caracter√≠sticas por importancia
    importance_df = importance_df.sort_values(by='Importance', ascending=False)
    
    return importance_df
```

```python
def seleccionar_caracter√≠sticas(importance_df: pd.DataFrame, df: pd.DataFrame, params: Dict[str, Any]) -> pd.DataFrame:
    """
    Selecciona las primeras 25 caracter√≠sticas m√°s importantes.

    Parameters
    ----------
    importance_df : pd.DataFrame
        DataFrame con la importancia de las caracter√≠sticas.
    df : pd.DataFrame
        DataFrame original con las caracter√≠sticas.
    params : Dict[str, Any]
        Diccionario de par√°metros que incluye las caracter√≠sticas requeridas.

    Returns
    -------
    pd.DataFrame
        DataFrame con la importancia de las caracter√≠sticas seleccionadas.
    """
    logger.info("Seleccionando las 25 caracter√≠sticas m√°s importantes...")

    # Asegurar que las caracter√≠sticas requeridas est√©n en importance_df
    required_features = [
        'sum_saldo_activos', 'sum_saldo_pasivos', 'delta_ah_sum_monto_trx_credito',
        'delta_ah_sum_monto_trx_debito', 'delta_tc_sum_facturacion_total', 
        'num_activos_prev', 'num_pasivos_prev', 'num_prod_activos', 'num_prod_pasivos'
    ]
    
    required_features_in_df = [feat for feat in required_features if feat in df.columns]
    
    for feat in required_features_in_df:
        if feat not in importance_df['Feature'].values:
            importance_df = pd.concat([importance_df, pd.DataFrame({'Feature': [feat], 'Importance': [0.001]})], ignore_index=True)
    
    # Ordenar por importancia y seleccionar las primeras 25 caracter√≠sticas
    importance_df = importance_df.sort_values(by='Importance', ascending=False).head(25)

    logger.info(f"Se seleccionaron {len(importance_df)} caracter√≠sticas importantes.")
    return importance_df

```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
import pandas as pd
import logging
from typing import Dict, Any
def procesar_importancia_pd(df: pd.DataFrame, params: Dict[str, Any]) -> pd.DataFrame:
    """
    Procesa la importancia de las caracter√≠sticas utilizando un modelo de regresi√≥n.

    Parameters
    ----------
    df : pd.DataFrame
        DataFrame con las caracter√≠sticas y la variable objetivo.
    params : Dict[str, Any]
        Diccionario de par√°metros que incluye la columna a conservar.

    Returns
    -------
    pd.DataFrame
        DataFrame con la importancia de las caracter√≠sticas ordenadas, asegurando un m√≠nimo de 25 caracter√≠sticas.
    """
    logger.info("Iniciando el proceso de c√°lculo de importancia de caracter√≠sticas...")

    def procesar_df(df: pd.DataFrame) -> pd.DataFrame:
        """
        Funci√≥n interna que procesa el DataFrame para calcular la importancia de las caracter√≠sticas.
        """
        # Identificar columnas categ√≥ricas y de fecha
        categorical_columns = df.select_dtypes(include=['object']).columns.tolist()
        # datetime_columns = df.select_dtypes(include=['datetime64']).columns.tolist()
        columna_conservar = params['columna_conservar']
        
        # Codificaci√≥n de columnas categ√≥ricas
        df_encoded = df.copy()
        label_encoders = {}
        for col in categorical_columns:
            le = LabelEncoder()
            df_encoded[col] = le.fit_transform(df_encoded[col].astype(str))
            label_encoders[col] = le

        # # Convertir columnas de fecha a num√©ricas (segundos desde 1970)
        # for col in datetime_columns:
        #     df_encoded[col] = (df_encoded[col] - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s')

        # Imputar valores faltantes por moda o mediana
        for col in df_encoded.columns:
            if df_encoded[col].dtype == 'object':
                # Imputaci√≥n por moda para columnas categ√≥ricas
                df_encoded[col].fillna(df_encoded[col].mode()[0], inplace=True)
            else:
                # Imputaci√≥n por mediana para columnas num√©ricas
                df_encoded[col].fillna(df_encoded[col].median(), inplace=True)

        # Convertir todos los datos a num√©ricos
        df_encoded = df_encoded.apply(pd.to_numeric, errors='coerce')

        # Separar caracter√≠sticas y variable objetivo
        X = df_encoded.drop(columns=[columna_conservar])
        y = df_encoded[columna_conservar]
        
       # Ajustar el modelo de regresi√≥n y calcular importancia
        regressor = DecisionTreeRegressor(max_depth=5, random_state=42)
        regressor.fit(X, y)
        
        importances = regressor.feature_importances_
        feature_names = X.columns
        importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
        importance_df = importance_df.sort_values(by='Importance', ascending=False)

        return importance_df

    importance_df = procesar_df(df)
    
    # Asegurar que al menos 25 caracter√≠sticas sean seleccionadas
    required_features = [
        'sum_saldo_activos', 'sum_saldo_pasivos', 'delta_ah_sum_monto_trx_credito',
        'delta_ah_sum_monto_trx_debito', 'delta_tc_sum_facturacion_total', 
        'num_activos_prev', 'num_pasivos_prev', 'num_prod_activos', 'num_prod_pasivos'
    ]
    
    # Filtrar required_features que existen en el DataFrame original
    required_features_in_df = [feat for feat in required_features if feat in df.columns]
    
    # Asegurar que las required_features est√©n en importance_df
    for feat in required_features_in_df:
        if feat not in importance_df['Feature'].values:
            importance_df = pd.concat([importance_df, pd.DataFrame({'Feature': [feat], 'Importance': [0.001]})], ignore_index=True)
    
    # Si despu√©s de agregar las required_features a√∫n hay menos de 25 caracter√≠sticas, a√±adir m√°s
    if len(importance_df) < 25:
        # Obtener todas las caracter√≠sticas restantes que no fueron seleccionadas
        remaining_features = df.drop(columns=params['columna_conservar']).columns.difference(importance_df['Feature'])
        # Crear un DataFrame con estas caracter√≠sticas restantes y asignarles una importancia baja
        additional_features = pd.DataFrame({
            'Feature': remaining_features,
            'Importance': [0.001] * len(remaining_features)  # Se asigna una importancia muy baja
        })
        # Concatenar con el importance_df y seleccionar las primeras 25
        importance_df = pd.concat([importance_df, additional_features]).head(25)

    logger.info(f"Se seleccionaron {len(importance_df)} caracter√≠sticas importantes.")
    return importance_df
```

```python
import shap
import xgboost
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier

def calcular_shap_values(X_train, y_train, info_save_select, **params):
    """
    Funci√≥n para calcular los SHAP values de un modelo basado en el tipo de modelo.
    
    Par√°metros:
    -----------
    X_train: pd.DataFrame
        DataFrame con los datos de entrenamiento.
    
    y_train: pd.DataFrame
        DataFrame con las etiquetas de entrenamiento.
    
    info_save_select: dict
        Diccionario que contiene la informaci√≥n del modelo. 
        Debe tener la clave 'modelo_produccion' -> 'best_model'.
    
    **params: dict
        Par√°metros adicionales opcionales.
        
    Retorna:
    --------
    shap_values: array
        Los valores SHAP calculados.
    """
    
    # Extraer el modelo del diccionario
    best_model = info_save_select['modelo_produccion']['best_model']
    
    # Inicializaci√≥n de variables
    explainer = None
    shap_values = None
    
    # Verificar si el modelo es una red neuronal MLP
    if isinstance(best_model, MLPClassifier):
        print("El modelo es un MLP. Utilizando Kernel SHAP o Deep SHAP.")
        
        # Para redes neuronales, podemos usar KernelExplainer o DeepExplainer (si aplica)
        explainer = shap.KernelExplainer(best_model.predict_proba, X_train)
        shap_values = explainer.shap_values(X_train)
    
    # Verificar si el modelo es XGBoost o RandomForest
    elif isinstance(best_model, xgboost.XGBClassifier) or isinstance(best_model, RandomForestClassifier):
        print("El modelo es un XGBoost o RandomForest. Utilizando TreeExplainer.")
        
        # Para modelos basados en √°rboles, TreeExplainer es la opci√≥n m√°s eficiente
        explainer = shap.TreeExplainer(best_model)
        shap_values = explainer.shap_values(X_train)
    
    else:
        raise ValueError("El modelo no es compatible con el c√°lculo de SHAP values.")

    # Retornar los valores SHAP calculados
    return shap_values
   
```

```python
import pandas as pd
from sklearn.experimental import enable_iterative_imputer  # Necesario para habilitar IterativeImputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import LabelEncoder
from typing import Dict, Any
import logging

# Configuraci√≥n del logger
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

def preprocesar_datos(df: pd.DataFrame, params: Dict[str, Any]) -> pd.DataFrame:
    """
    Preprocesa el DataFrame codificando variables categ√≥ricas y transformando fechas, 
    diferenciando por tipos de variables y utilizando t√©cnicas avanzadas de imputaci√≥n.

    Parameters
    ----------
    df : pd.DataFrame
        DataFrame con las caracter√≠sticas y la variable objetivo.
    params : Dict[str, Any]
        Diccionario de par√°metros que incluye la columna a conservar y tipos de variables.

    Returns
    -------
    pd.DataFrame
        DataFrame preprocesado con columnas categ√≥ricas codificadas, fechas transformadas 
        y valores imputados.
    """
    # Identificar columnas por tipo especificado en params
    socio_demographic_vars = params.get('socio_demographic_vars', [])
    transactional_vars = params.get('transactional_vars', [])
    product_related_vars = params.get('product_related_vars', [])
    columna_conservar = params.get('columna_conservar', None)

    # Identificar columnas que no est√°n en ninguna de las tres clases ni en la columna a conservar
    all_defined_columns = socio_demographic_vars + transactional_vars + product_related_vars + [columna_conservar]
    additional_columns = df.columns.difference(all_defined_columns).tolist()

    # Log de columnas adicionales
    if additional_columns:
        logger.info(f"Las columnas adicionales no categorizadas son: {additional_columns}")

    # Codificaci√≥n de variables categ√≥ricas
    categorical_columns = df.select_dtypes(include=['object']).columns.tolist()
    datetime_columns = df.select_dtypes(include=['datetime64']).columns.tolist()

    # Codificaci√≥n de columnas categ√≥ricas
    df_encoded = df.copy()
    label_encoders = {}
    for col in categorical_columns:
        le = LabelEncoder()
        df_encoded[col] = le.fit_transform(df_encoded[col].astype(str))
        label_encoders[col] = le

    # Convertir columnas de fecha a num√©ricas (segundos desde 1970)
    for col in datetime_columns:
        df_encoded[col] = (df_encoded[col] - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s')

    # Aplicar imputaci√≥n avanzada por tipo de variable
    # Imputaci√≥n de variables socio-demogr√°ficas (mediana, ya que puede haber sesgos)
    if socio_demographic_vars:
        df_encoded[socio_demographic_vars] = df_encoded[socio_demographic_vars].apply(lambda x: x.fillna(x.median()))

    # Imputaci√≥n de variables transaccionales (usando Iterative Imputer)
    if transactional_vars:
        imputer = IterativeImputer(random_state=42)
        df_encoded[transactional_vars] = imputer.fit_transform(df_encoded[transactional_vars])

    # Imputaci√≥n de variables relacionadas con el producto (usando Iterative Imputer)
    if product_related_vars:
        imputer = IterativeImputer(random_state=42)
        df_encoded[product_related_vars] = imputer.fit_transform(df_encoded[product_related_vars])

    # Imputaci√≥n de columnas adicionales no categorizadas
    for col in additional_columns:
        if df_encoded[col].dtype == 'object':
            # Imputaci√≥n para variables categ√≥ricas con la moda
            df_encoded[col].fillna(df_encoded[col].mode()[0], inplace=True)
        else:
            # Imputaci√≥n para variables num√©ricas con la mediana o utilizando IterativeImputer
            imputer = IterativeImputer(random_state=42)
            df_encoded[[col]] = imputer.fit_transform(df_encoded[[col]])

    # Verificar si la columna a conservar tiene valores nulos
    if columna_conservar and df_encoded[columna_conservar].isnull().any():
        logger.info(f"La variable objetivo {columna_conservar} contiene valores nulos.")

    # Log de finalizaci√≥n de preprocesamiento
    logger.info("El preprocesamiento del DataFrame ha sido exitoso.")

    return df_encoded, label_encoders

```

```python
def calculate_new_variables(df: pd.DataFrame) -> pd.DataFrame:
    """
    Calcula nuevas variables basadas en la suma de saldos de activos y pasivos,
    as√≠ como los deltas entre diferentes periodos para columnas espec√≠ficas.

    Parameters
    ----------
    df : pd.DataFrame
        DataFrame con las columnas necesarias para calcular las nuevas variables.

    Returns
    -------
    pd.DataFrame
        DataFrame con las nuevas variables calculadas.
    """
    logger.info("Calculando nuevas variables de saldos y deltas...")

    df['sum_saldo_activos'] = (
        df['ah_sum_saldo_promedio'] +
        df['cc_sum_saldo_promedio'] +
        df['cdt_sum_monto_apertura'] +   
        df['cdt_sum_valor_interes']
    )
    df['sum_saldo_pasivos'] = (
        df['tc_sum_saldo_mean_2'] +
        df['ld_sum_saldo_promedio'] +
        df['lib_saldo_promedio'] +
        df['sus_saldo_promedio'] +
        df['veh_sum_saldo_promedio'] +
        df['viv_sum_saldo_promedio']
    )
    df['delta_ah_sum_monto_trx_credito'] = df['ah_sum_monto_trx_credito_mean_2'] - df['ah_sum_monto_trx_credito_mean_6']
    df['delta_ah_sum_monto_trx_debito'] = df['ah_sum_monto_trx_debito_mean_2'] - df['ah_sum_monto_trx_debito_mean_6']
    df['delta_tc_sum_facturacion_total'] = df['tc_sum_facturacion_total_valor_mean_2'] - df['tc_sum_facturacion_total_valor_mean_6']
    
    logger.info("C√°lculo de nuevas variables completado.")
    return df
```

```python
import pandas as pd
from xgboost import XGBRegressor
from typing import Dict, Any

def procesar_importancia_pd(df: pd.DataFrame, params: Dict[str, Any]) -> pd.DataFrame:
    """
    Procesa la importancia de las caracter√≠sticas utilizando un modelo de regresi√≥n.

    Parameters
    ----------
    df : pd.DataFrame
        DataFrame con las caracter√≠sticas y la variable objetivo.
    params : Dict[str, Any]
        Diccionario de par√°metros que incluye la columna a conservar.

    Returns
    -------
    pd.DataFrame
        DataFrame con la importancia de las caracter√≠sticas ordenadas.
    """
    logger.info("Iniciando el proceso de c√°lculo de importancia de caracter√≠sticas...")

    def procesar_df(df: pd.DataFrame) -> pd.DataFrame:
        """
        Funci√≥n interna que procesa el DataFrame para calcular la importancia de las caracter√≠sticas.
        """
        # Identificar columnas categ√≥ricas y de fecha
        categorical_columns = df.select_dtypes(include=['object']).columns.tolist()
        datetime_columns = df.select_dtypes(include=['datetime64']).columns.tolist()
        columna_conservar = params['columna_conservar']
        
        # Imputar valores faltantes por moda o mediana
        for col in categorical_columns:
            df[col].fillna(df[col].mode()[0], inplace=True)
        
        for col in df.columns:
            if df[col].dtype != 'object':  # Para columnas no categ√≥ricas
                df[col].fillna(df[col].median(), inplace=True)

        # Convertir columnas de fecha a num√©ricas (segundos desde 1970)
        for col in datetime_columns:
            df[col] = (df[col] - pd.Timestamp("1970-01-01")) // pd.Timedelta('1s')

        # Convertir columnas categ√≥ricas a tipo 'category' para que XGBoost las maneje internamente
        for col in categorical_columns:
            df[col] = df[col].astype('category')

        # Convertir todos los datos a num√©ricos (despu√©s de la conversi√≥n de fechas y categ√≥ricas)
        df_encoded = df.apply(pd.to_numeric, errors='coerce')

        # Separar caracter√≠sticas y variable objetivo
        X = df_encoded.drop(columns=[columna_conservar, 'periodo', 'hashvalue1'], errors='ignore')
        y = df_encoded[columna_conservar]
        
        # Ajustar el modelo de regresi√≥n con XGBoost y calcular importancia
        regressor = XGBRegressor(n_estimators=40, max_depth=5, random_state=42, verbosity=0, enable_categorical=True)
        regressor.fit(X, y)
        
        importances = regressor.feature_importances_
        feature_names = X.columns
        importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
        importance_df = importance_df[importance_df['Importance'] > 0].sort_values(by='Importance', ascending=False)
        
        return importance_df

    importance_df = procesar_df(df)
    logger.info(f"Se seleccionaron {len(importance_df)} caracter√≠sticas importantes.")
    return importance_df
```
