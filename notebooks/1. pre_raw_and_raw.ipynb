```python
from typing import Dict, List, Any, Tuple
import polars as pl
```

```python
def concat_dataframes_pl_pd_borrar(
    df1: pl.DataFrame,
    df2: pl.DataFrame,
    df3: pl.DataFrame,
    df4: pl.DataFrame,
    df5: pl.DataFrame,
    df6: pl.DataFrame,
    df7: pl.DataFrame,
    df8: pl.DataFrame,
    df9: pl.DataFrame,
    df10: pl.DataFrame,
    df11: pl.DataFrame,
    df12: pl.DataFrame,
    df13: pl.DataFrame,
    params: Dict[str, Any]
) -> pd.DataFrame:

    variables = params['vars']
    

    def convertir_a_float64(df: pl.DataFrame, cols: list) -> pl.DataFrame:
        gc.collect()
        try:
            # si es un parquet pasarlo a polars
            df = pl.from_pandas(df)
        except:
            pass
        numeric_cols = [col for col in cols if df[col].dtype in [pl.Float64, pl.Int64, pl.Int32, pl.Float32]]
        ret = df.select(cols).with_columns([pl.col(col).cast(pl.Float64) for col in numeric_cols])
        gc.collect()
        return ret
        
    logger.info("Iniciando el proceso de filtrado de columnas ...")
    gc.collect()
    df1 = convertir_a_float64(df1, variables)
    df2 = convertir_a_float64(df2, variables)
    df3 = convertir_a_float64(df3, variables)
    df4 = convertir_a_float64(df4, variables)
    df5 = convertir_a_float64(df5, variables)
    df6 = convertir_a_float64(df6, variables)
    df7 = convertir_a_float64(df7, variables)
    df8 = convertir_a_float64(df8, variables)
    df9 = convertir_a_float64(df9, variables)
    df10 = convertir_a_float64(df10, variables)
    df11 = convertir_a_float64(df11, variables)
    df12 = convertir_a_float64(df12, variables)
    df13 = convertir_a_float64(df13, variables)
    gc.collect()
    logger.info("Iniciando el proceso de concatenación de DataFrames...")
    df_list = [df1, df2, df3, df4, df5, df6, df7, df8, df9,  df10, df11, df12, df13]

    df_final = pl.concat(df_list, how='vertical')
    gc.collect()
    logger.info(f"DataFrame concatenado: {df_final.shape[0]} registros y {df_final.shape[1]} columnas.")

    df_final_pd = df_final.to_pandas()

    if df_final_pd is None or df_final_pd.empty:
        logger.error("El DataFrame final es None o está vacío.")
        return None

    logger.info("Proceso de concatenación completado y convertido a DataFrame de pandas.")
    return df_final_pd
```

## caracteres con mas información: pre_raw parameters

```python
pct_min_nulos = 0.1
```

```python
varss_want = []
for base in ["master_ago2023","master_sep2023","master_oct2023","master_nov2023","master_dic2023",
             "master_ene2024","master_feb2024","master_mar2024", "master_abril2024","master_mayo2024",
             "master_junio2024","master_julio2024","master_agosto2024"]:
    print(base)
    master_MMAAAA = catalog.load(base)
    try:
        master_MMAAAA = master_MMAAAA.to_pandas()
    except:
        pass
    nulls = master_MMAAAA.isnull().sum()
    nulos = pd.concat([nulls.sort_values(ascending=True),nulls.sort_values(ascending=True)/master_MMAAAA.shape[0]], axis = 1)
    tot_ = nulos.shape[0]
    nulos = nulos[nulos[1]<=pct_min_nulos]
    print('N Variables admisibles',nulos.shape[0], 'de ', tot_)
    varss_want = list(set(list(varss_want+ nulos.index.tolist())))
```

**Ordenando la lista**

```python
dicct = {'ids': ['hashvalue1','periodo']}
filt_order = []
for col in varss_want:
    if col not in dicct['ids']:
        filt_order.append(col)
for col in varss_want:
    if len(col.split('_')) >1:
        if  col.split('_')[0] not in list(dicct.keys()):
            dicct[col.split('_')[0]] = [col]
        else:
            dicct[col.split('_')[0]] = dicct[col.split('_')[0]]+[col]
    else:
        dicct['ids'] = dicct['ids']+[col]

### uniendo las variables unicas en ids:
remove = []
for col in dicct.keys():
    if len(dicct[col])==1:
        remove.append(col)
        dicct['ids'] = dicct['ids'] +dicct[col]
for col in remove:
    del dicct[col]
dicct.keys()
```

```python
order_columns_want = []
for col in dicct.keys():
    order_columns_want = order_columns_want + dicct[col]
# elimino dupplicados sin perder el orden
order_columns_want = pd.DataFrame(order_columns_want).drop_duplicates().values.flatten().tolist()
len(order_columns_want)
```
